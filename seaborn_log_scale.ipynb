## DJI index (monthly)

```python
df=pd.read_csv('https://stooq.com/q/d/l/?s=^dji&i=m', parse_dates=['Date'], index_col='Date')
df
```

```python
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
plt.figure(figsize=(10, 7))
sns.lineplot(df.index, df.Close)
plt.title('DJIA')
plt.grid()
```

```python
fig, ax = plt.subplots(figsize=(10, 7))
ax.set(yscale="log")
sns.lineplot(df.index, df.Close, ax=ax)
plt.title('DJIA')
plt.grid()
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
sns.lineplot(df.index, df.Close, ax=ax, color='orange', label='standard')

ax2=ax.twinx()

ax2.set(yscale="log")
sns.lineplot(df.index, df.Close, ax=ax2, label='log')

plt.title('DJIA')
fig.legend()
```
