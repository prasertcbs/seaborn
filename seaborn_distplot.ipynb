```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
print(pd.__version__)
print(sns.__version__)
```

```python
df=pd.read_csv('https://github.com/prasertcbs/tutorial/raw/master/mpg.csv')
df.head()
```

```python
sns.distplot(df['cty'])
```

```python
sns.distplot(df['cty'], kde=False)
```

```python
sns.distplot(df['cty'], kde=False)
plt.axvline(x=df.cty.mean(), color="orange", linestyle="--")
plt.title(f"mean = {df.cty.mean():.2f}, sd = {df.cty.std():.2f}")
```

```python
sns.distplot(df['cty'], kde=True, label="city")
sns.distplot(df['hwy'], kde=True, label="highway")
plt.legend()
```

```python
sns.distplot(df['cty'], kde=True, label="city")
plt.show()
sns.distplot(df['hwy'], kde=True, label="highway")
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, sharex=True, sharey=True)
print(ax.shape)
sns.distplot(df['cty'], kde=False, ax=ax[0])
sns.distplot(df['hwy'], kde=False, ax=ax[1])
```
