# Harvard Tuition from 1985-2017
Tuition data from Harvard's College and graduate/professional schools since 1985  
data source: https://www.kaggle.com/harvard-university/harvard-tuition/data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
print(f'pandas  version: {pd.__version__}')
print(f'seaborn version: {sns.__version__}')
```

```python
url='https://github.com/prasertcbs/basic-dataset/raw/master/tuition_graduate.csv'
df=pd.read_csv(url)
df.head()
```

```python
df.tail()
```

```python
df.school.unique()
```

```python
sns.lineplot(data=df,
             x='academic.year', y='cost', 
             hue='school')
```

```python
sns.lineplot(data=df[df.school.str.match('Business|Design|Law')],
             x='academic.year', y='cost', 
             style='school',
             hue='school')
```

```python
ax=sns.lineplot(data=df[df.school.str.match('Business|Design|Law')],
             x='academic.year', y='cost', 
             style='school',
             hue='school')
y_values=ax.get_yticks()
ax.set_yticklabels([f'{y:,.0f}' for y in y_values]);
```

```python
sns.relplot(data=df[df.school.str.match('Business|Design|Law')],
             x='academic.year', y='cost', 
             style='school',
             col='school'
             )
```

```python
sns.set_context('notebook', font_scale=1.6)
sns.relplot(data=df,
             x='academic.year', y='cost', 
             kind='line',
             hue='school',
             col='school',
             col_wrap=5,
             )
```
