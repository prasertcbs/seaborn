```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
print(f"pandas  version = {pd.__version__}")
print(f"seaborn version = {sns.__version__}")
```

## read miles per gallon dataset (mpg.csv)

```python
url='https://github.com/prasertcbs/tutorial/raw/master/mpg.csv'
df=pd.read_csv(url)
df.head()
```

```python
df['gear']=df.trans.str[0].map({'a':'auto', 'm':'manual'}).astype("category")
df['drv']=df.drv.map({'f':'front', 'r':'rear', '4':'4-wheel'}).astype('category')
df.head()
```

```python
sns.barplot(x='year', y='cty', data=df,
               ci=None)
```

### add hue

```python
sns.barplot(x='year', y='cty', data=df,
               hue='gear',
               ci=None)
```

### use catplot/factorplot

```python
sns.catplot(x='year', y='cty', data=df,
               hue='gear', 
               kind='bar',
               ci=None,
               height=4, aspect=1)
```

### add 'drv' as column factor

```python
sns.catplot(x='year', y='cty', data=df,
               hue='gear', 
               kind='bar',
               ci=None,
               col='drv',
               height=4, aspect=1)
```

```python
df['class'].unique()
```

```python
sns.catplot(x='year', y='cty', data=df,
               hue='gear', 
               kind='bar',
               ci=None,
               col='drv',
               row='class',
               height=4, aspect=1)
```

```python
df[df['class'].str.match('^(compact|midsize|suv)')].sample(frac=.05)
```

### add 'class' as row factor

```python
sns.catplot(x='year', y='cty', data=df[df['class'].str.match('compact|midsize|suv')],
               hue='gear', 
               kind='bar',
               ci=None,
               row='class',
               col='drv',
               height=3, aspect=1)
```

### use point chart (kind)

```python
sns.catplot(x='year', y='cty', data=df,
               hue="gear", 
               col="drv",
               ci=None,
               height=4, aspect=1);
```

### col_wrap

```python
sns.catplot(x='year', y='cty', data=df,
               hue="gear", 
               col="class",
               ci=None,
               height=2.5, aspect=1);
```

```python
sns.catplot(x='year', y='cty', data=df,
               hue="gear", 
               col="class",
               col_wrap=3,
               ci=None,
               height=2.5, aspect=1);
```

### kind = 'violin'

```python
sns.catplot(x='year', y='cty', data=df,
               hue='gear', 
               kind='violin',
               col='class',
               col_wrap=4,
               ci=None,
               height=2.5, aspect=1);
```

```python
sns.catplot(x='year', y='cty', data=df,
               hue='gear', 
               kind='box',
               col='class',
               col_wrap=4,
               ci=None,
               height=2.5, aspect=1);
```

```python
sns.catplot(x='year', y='cty', data=df,
               hue="gear", 
               kind="bar",
               col="class",
               col_wrap=4,
               height=2.5, aspect=1);
```

### named palette

```python
sns.catplot(x='drv', y='cty', data=df,
               hue="gear", 
               palette='Set2',
               col="year",
               kind="box",
               height=4, aspect=1);
```

```python
sns.catplot(x='gear', y='cty', data=df,
               hue="drv", 
               palette='Set2',
               col="year",
               kind="box",
               height=4, aspect=1);
```

### palette color

```python
sns.catplot(x='drv', y='cty', data=df,
               hue="gear", 
               palette=['deepskyblue', 'tomato'],
               col="year",
               kind="box",
               height=4, aspect=1);
```

### palette dict + hex color #rrggbb

```python
sns.catplot(x='drv', y='cty', data=df,
               hue="gear", 
               palette={'auto':'deepskyblue', 'manual':'#CC0000'},
               col="year",
               kind="box",
               height=4, aspect=1);
```
