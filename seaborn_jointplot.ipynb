```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
print(f"pandas  version = {pd.__version__}")
print(f"seaborn version = {sns.__version__}")
```

```python
df=pd.read_csv('https://github.com/prasertcbs/tutorial/raw/master/mpg.csv')
df.head()
```

```python
sns.distplot(df['displ']);
```

```python
sns.distplot(df['hwy']);
```

```python
fig, ax = plt.subplots(1, 2, sharex=False, sharey=False, figsize=(9, 4))
print(ax.shape)
sns.distplot(df['displ'], kde=False, ax=ax[0])
sns.distplot(df['hwy'], kde=False, ax=ax[1]);
```

```python
sns.regplot(data=df, x='displ', y='hwy');
```

```python
sns.jointplot(data=df, x='displ', y='hwy');
```

```python
sns.jointplot(data=df, x='displ', y='hwy', kind='reg');
```

```python
sns.jointplot(data=df, x='displ', y='hwy', kind='hex');
```

```python
sns.jointplot(data=df, x='displ', y='hwy', kind='kde');
```

```python
sns.jointplot(data=df, x='displ', y='hwy', kind='reg', 
              color='orange',
              size=5, ratio=3);
```

```python
sns.jointplot(data=df, x='displ', y='hwy', kind='reg', 
              color='orange',
              size=5, ratio=2);
```

```python
sns.jointplot(data=df, x='displ', y='hwy', kind='reg', xlim=(1, 3));
```

```python
sns.jointplot(data=df, x='displ', y='hwy', kind='reg', ylim=(30, 50));
```

```python
sns.jointplot(data=df, x='displ', y='hwy', kind='reg', xlim=(0, 4), ylim=(30, 50));
```
